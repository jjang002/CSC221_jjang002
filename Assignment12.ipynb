```python
# Jaeyoung Jang
class Money:
    def __init__(self, value):
        self.value = value

    def __str__(self):
        return f"${self.value:.2f}"

# Define the child class Bills
class Bills(Money):
    pass

# Define child classes for specific bill denominations
class One(Bills):
    def __init__(self):
        super().__init(1.00)

class Five(Bills):
    def __init__(self):
        super().__init(5.00)

class Ten(Bills):
    def __init__(self):
        super().__init(10.00)

class Twenty(Bills):
    def __init__(self):
        super().__init(20.00)

# Define the child class Coins
class Coins(Money):
    pass

# Define child classes for specific coin denominations
class Penny(Coins):
    def __init__(self):
        super().__init(0.01)

class Nickel(Coins):
    def __init__(self):
        super().__init(0.05)

class Dime(Coins):
    def __init__(self):
        super().__init(0.10)

class Quarter(Coins):
    def __init__(self):
        super().__init(0.25)

# Function to calculate the total value of a list of Money objects
def calculate_total(money_list):
    total = sum(money.value for money in money_list)
    return total

# Function to print a list of Money objects
def print_money(money_list):
    for money in money_list:
        print(money)

# Test cases
if __name__ == "__main__":
    # Create wallets with bills and coins
    wallet1 = [One(), One(), Five(), Quarter(), Dime()]
    wallet2 = [Ten(), Ten(), Penny(), Nickel()]

    # Calculate and print the total value of each wallet
    total_wallet1 = calculate_total(wallet1)
    total_wallet2 = calculate_total(wallet2)

    print("Wallet 1:")
    print_money(wallet1)
    print(f"Total value: {total_wallet1:.2f}\n")

    print("Wallet 2:")
    print_money(wallet2)
    print(f"Total value: {total_wallet2:.2f}")
```
